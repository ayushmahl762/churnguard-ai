```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
sns.countplot(x='churned', data=customers)

plt.title('Churn Distribution')
plt.show()
```

```python
sns.histplot(customers['total_spend_usd'], bins=30)

plt.title('Distribution of Total Spend')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='churned',
    y='days_since_last_purchase',
    data=customers
)

plt.title('Days Since Last Purchase vs Churn')

plt.show()
```
